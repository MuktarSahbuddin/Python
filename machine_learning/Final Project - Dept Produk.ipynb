```python
#Membuat kolom baru
def kesimpulan (row):
   if row['lama_produk_terjual'] < 31 :
      return 'Diminati'
   if row['lama_produk_terjual'] > 30:
      return 'Cukup Diminati'
   if row['lama_produk_terjual'] > 90:
      return 'Kurang Diminati'
   if row['lama_produk_terjual']  == 0:
      return 'Tidak Diminati'
   return 'Tidak Diminati'
```

Dapat dilihat bahwa kolom yang memiliki outlier adalah cost, product retail price dan lama produk terjual.

Selanjutnya, kita akan menggunakan IQR untuk menghapus outlier.

```python
#Menghapus outlier pada kolom cost
Q1 = fix_product["cost"].quantile(0.25)
Q3 = fix["cost"].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (IQR * 1.5)
high_limit = Q3 + (IQR * 1.5)
```

```python
fix_product = fix_product[(fix_product["cost"] >= low_limit) &
                        (fix["cost"] <= high_limit)
                        ]
```

```python
print("cost", fix_product.shape)
```

```python
#Menghapus outlier pada kolom product retail price
Q1 = fix_product["product_retail_price"].quantile(0.25)
Q3 = fix_product["product_retail_price"].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (IQR * 1.5)
high_limit = Q3 + (IQR * 1.5)
```

```python
fix_product = fix_product[(fix_product["product_retail_price"] >= low_limit) &
                        (fix["product_retail_price"] <= high_limit)
                        ]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
x = fix_product.drop(columns='Hasil')
y = fix_product['Hasil']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
tree_product = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=1)
tree_product.fit(xtrain,ytrain)
```

```python
ypred = tree_product.predict(xtest)
cm = confusion_matrix(ytest, ypred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, ypred)
print("Tingkat Akurasi Algoritma C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(ytest, ypred)
print("Tingkat Akurasi: %d persen" %(akurasi*100))
```
