```python
# Import Package
import pandas as pd # Data Manipulating
import numpy as np # Support large multi-dimensional array
import matplotlib.pyplot as plt #Basic Plot di Python
import seaborn as sns #plot alternatif dari matplotlib dengan beberapa fitur berebeda
pd.options.mode.chained_assignment = None  # default='warn
pd.set_option('display.float_format', lambda x: '%.2f' % x) # set 2 angka dibelakang koma (desimal)
```

```python
# Import data users dan events
data_users = pd.read_csv('/content/drive/MyDrive/users.csv')
data_events = pd.read_csv('/content/drive/MyDrive/events.csv')
```

```python
# Memanggil data
data_users
```

```python
# Memanggil data 
data_events
```

```python
# Memilih salah satu kolom 
data_users = data_users[['id','first_name','gender']] 
data_users[['id','first_name','gender']] #Memanggil data
```

```python
data_events = data_events[['user_id','created_at','event_type']]
data_events[['user_id','created_at','event_type']]#Memanggil data
```

```python
left_merge = pd.merge(data_users, data_events, how ='left', left_on='id', right_on='user_id') #LEFT JOIN USER DAN EVENT
left_merge.head(1000) #TAMPILIN DATA SEJUMLAH 1000
```

```python
left_merge[left_merge.isna().any(axis=1)] #Data Yang NA
```

```python
left_merge.shape #melihat jumlah data sebelum drop NA
```

```python
left_merge=left_merge.dropna() #menghapus data NA
left_merge.shape
```

```python
left_merge[left_merge.isna().any(axis=1)] #check data sudah di drop yang NA
```
