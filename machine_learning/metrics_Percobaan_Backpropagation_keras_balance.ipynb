```python
# Import all libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/Client/blabla/Final_Putusan_Perceraian.xlsx')
df.head()
```

```python
df = df.drop(['No', 'Mediasi'], axis=1)
df.head()
```

```python
df.Putusan.value_counts()
```

```python
Dikabulkan= df[df.Putusan == 'Dikabulkan'][:100]
Ditolak= df[df.Putusan == 'Ditolak']
df = pd.concat([Dikabulkan,Ditolak])
```

```python
#Transormasi Data menggunakan LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Putusan']= le.fit_transform(df['Putusan'].values)
df['Jenis']= le.fit_transform(df['Jenis'].values)
```

```python
df
```

```python
df.Putusan.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop(columns='Putusan').values
y = df['Putusan'].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

##Import minmaxscaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python

import tensorflow as tf


model = tf.keras.models.Sequential() #arsitektur

model.add(tf.keras.layers.Dense(6,input_shape=[6] ))
model.add(tf.keras.layers.Dense(9, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))

model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.09), metrics=['accuracy'])
```

```python
weight=model.get_weights()
weight
```

```python
model.fit(X_train,y_train, epochs=300, batch_size=128)
```

```python
val_loss,val_acc = model.evaluate(X_test, y_test, batch_size=32)
print(val_loss,val_acc)
```

```python
predict_test = model.predict(X_test).round()
print(predict_test)
```
