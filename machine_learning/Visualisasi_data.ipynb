```python
data_transformasi.loc[data_transformasi["gender"] == "M", "gender"] = 1
```

```python
print("\n[->] Data eksplorasi dengan dengan mengecek korelasi dari setiap feature Data Transformasi")
print(data_transformasi.corr())
```

Analisa dengan Heatmap

```python
# HeatMap Transformasi
Transformasi_map = data_transformasi.pivot_table("id", "user_id","gender")
ax = sns.heatmap(Transformasi_map,cmap="YlGnBu")
```

```python
#pairplot() dari seaborn (sns) modul untuk menggambarkan hubungan setiap feature Data Transformasi.
plt.figure()
sns.pairplot(data_transformasi)
plt.show()
```

Analisa Outlier Data

```python
#Cek Data Outlier pada Data Transfromasi
ax = sns.boxplot(x="gender", y="user_id", data=data_transformasi)
```

```python
#calculate z-scores of `data_transformasi` to Remove data outlier
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(data_transformasi['gender']))
print(z)
```

dari Tampilan box plot dan user_id tidak terlihat adanya outlier di kedua variable tersebut
