```python
# df (DataFrame). Membaca dataset
df = pd.read_excel('/content/drive/MyDrive/Client/bla/Final.xlsx') 
# Menampilkan  dataset df yang paling atas sebanyak 5
df.head()
```

```python
# Ngedrop kolom no, mediasi, axis 1 itu menunjukan dia kolom
df = df.drop(['No', 'Mediasi'], axis=1) 
df.head()
```

```python
# Menampilkan jumlah kolom putusan dengan kelas dikabulkan sebanyak 724 dan kelas ditolak sebanyak 55
df.Putusan.value_counts()
```

```python
# Membuat kolom putusan dengan kelas dikabulkan menjadi 100 dan disimpan kedalam variabel Dikabulkan
Dikabulkan= df[df.Putusan == 'Dikabulkan'][:100] 
# Membuat kolom putusan dengan kelas Ditolak dan disimpan kedalam variabel Ditolak
Ditolak= df[df.Putusan == 'Ditolak']
# Menggabungkan  variabel Dikabulkan, Ditolakdan disimpan kedalam variabel df
df = pd.concat([Dikabulkan,Ditolak])
```

```python
#Transormasi Data menggunakan LabelEncoder
# Improt LabelEncoder
from sklearn.preprocessing import LabelEncoder 
# Memanggil fungsi  LabelEncoder dan menyimpan kedalam variabel le
le = LabelEncoder()
```

```python
# Membuat kolom putusan menjadi angka dengan angka 0 atau 1
df['Putusan']= le.fit_transform(df['Putusan'].values)
# Membuat kolom jenis menjadi angka dengan angka 0 atau 1 
df['Jenis']= le.fit_transform(df['Jenis'].values)
```

```python
df
```

```python
# Menampilkan jumlah kelas dengan angka 0 adalah kelas dikabulkan sebanyak 100 dan angka 1 adalah kelas ditolak
df.Putusan.value_counts() 
```

```python
# Import train_test_split untuk membagi data train & test
from sklearn.model_selection import train_test_split 
# Import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 
```

```python
# Menyimpan kolom x dengan hanya nge-drop kolom putusan yang tinggal kolom Jenis,UP,UT,UK,JA,Alasan
X = df.drop(columns='Putusan').values 
# Menyimpan kolom y dengan kolom putusan
y = df['Putusan'].values  
```

```python
# X_train untuk menampung data source yang akan dilatih. 
# X_test untuk menampung data target yang akan dilatih.
# y_train untuk menampung data source yang akan digunakan untuk testing
# y_test untuk menampung data target yang akan digunakan untuk testing

# Split data dan membuat jadi data train dan data test
# test_size atau pembagian data %
# stratify untuk mengacak data agar yang diambil tidak data dikabulkan atau ditolak
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42) 

# Menampilkan kolom dan baris untuk setiap X_train, X_test,y_train
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

Import MinMax Scalar

```python
# Improt MinMaxScaler
from sklearn.preprocessing import MinMaxScaler 
```

```python
# Memanggil fungsi  MinMaxScaler dan menyinpan ke dalam variabel scaler
scaler = MinMaxScaler() 
```

```python
# Menormalisasikan x_train
X_train = scaler.fit_transform(X_train)
# Menormalisasikan x_test
X_test = scaler.fit_transform(X_test)
```

```python
 # Menyimpan  x_train ke dalam data frame
X_train = pd.DataFrame(X_train)
X_train
```

```python
# Menyimpan  X_test ke dalam data frame
X_test = pd.DataFrame(X_test) 
X_test.head()
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential() 

model.add(tf.keras.layers.Dense(6,input_shape=[6] )) 
model.add(tf.keras.layers.Dense(12, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid')) 

model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.09), metrics=['accuracy'])
```

```python
# Model di training  dengan epoch 300 dan batch_size 32
model.fit(X_train,y_train, epochs=300, batch_size=128) 
```

```python
# Melakukan evaluasi model dan menghasilkan value loss dan value akurasi
val_loss,val_acc = model.evaluate(X_test, y_test, batch_size=32) 
print(val_loss,val_acc)
```

```python
# Model prediksi
predict_test = model.predict(X_test).round()
print(predict_test)
```

```python
# Membuat list klasifikasi dari precision,recall,f1-score, akurasi
print(classification_report(y_test, predict_test)) 
```
